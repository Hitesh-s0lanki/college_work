```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = [[1], [2], [3], [4]]
y = [2, 4, 6, 8]

# Model
model = LinearRegression()
model.fit(X, y)

# Plot
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
x = [[1], [2], [3], [4]]
y = [5, 9, 2, 8]

# Plot the Data in the Graph
plt.scatter(x, y, color='blue', label='Data')
plt.show()


model = LinearRegression()
model.fit(x, y)

# Plot the Data in the Graph
plt.scatter(x, y, color='blue', label='Data')   
plt.plot(x, model.predict(x), color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Sample time data (e.g., time steps)
time = np.arange(10).reshape(-1, 1)

# Binary target: 0 or 1
target = [0, 0, 0, 1, 0, 1, 1, 1, 1, 1]

# Logistic Regression
model = LogisticRegression()
model.fit(time, target)

# Predict probabilities
probs = model.predict_proba(time)[:, 1]

# Plot
plt.plot(time, target)
# plt.plot(time, probs, 'r--', label='Predicted Probability')
plt.xlabel("Time")
plt.ylabel("Value (0/1)")
plt.title("Logistic Regression")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Sample data
X = np.array([[1,2], [1,4], [1,0],
              [10,2], [10,4], [10,0]])

# Apply KMeans
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Plot
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.scatter(centers[:,0], centers[:,1], c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```
